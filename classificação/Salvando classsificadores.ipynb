```python
import pandas as pd

base = pd.read_csv('../data/credit_data.csv')
base.loc[base.age < 0, 'age'] = 40.92
               
features_ = base.iloc[:, 1:4].values
class_ = base.iloc[:, 4].values
```

```python
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = 'NaN', strategy = 'mean', axis = 0)
imputer = imputer.fit(features_[:, 1:4])
features_[:, 1:4] = imputer.transform(features_[:, 1:4])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features_= scaler.fit_transform(features_)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier

classifier_SVM = SVC(kernel = 'rbf', C = 2.0, probability = True)
classifier_SVM.fit(features_, class_)

classifier_RandomForest = RandomForestClassifier(n_estimators = 40, criterion = 'entropy')
classifier_RandomForest.fit(features_, class_)

classifier_MLP = MLPClassifier(verbose = True, max_iter = 1000,
                                 tol = 0.000010, solver = 'adam',
                                 hidden_layer_sizes=(100), activation = 'relu',
                                 batch_size = 200, learning_rate_init = 0.001)
classifier_MLP.fit(features_, class_)
```

```python
import pickle
pickle.dump(classifier_SVM, open('svm_finalizado.sav', 'wb'))
pickle.dump(classifier_RandomForest, open('random_forest_finalizado.sav', 'wb'))
pickle.dump(classifier_MLP, open('mlp_finalizado.sav', 'wb'))
```
